--------
Read CSV files
--------
--------

```python
import pandas as pd
dataset_dir = "../olist_dataset"

orders = pd.read_csv(f"{dataset_dir}/olist_orders_dataset.csv")
order_items = pd.read_csv(f"{dataset_dir}/olist_order_items_dataset.csv")
products = pd.read_csv(f"{dataset_dir}/olist_products_dataset.csv")
customers = pd.read_csv(f"{dataset_dir}/olist_customers_dataset.csv")
payments = pd.read_csv(f"{dataset_dir}/olist_order_payments_dataset.csv")

print(orders.shape)
print(order_items.shape)
print(products.shape)
print(customers.shape)
print(payments.shape)
```

--------
Explore columns
--------
--------

```python
print(orders.columns)
print(order_items.columns)
print(payments.columns)
print(customers.columns)
print(products.columns)
```

--------
Filter columns
--------
--------

```python
orders = orders[[
    "order_id",
    "customer_id",
    "order_purchase_timestamp"
]]

order_items = order_items[[
    "order_id",
    "product_id",
    "price"
]]

payments = payments[[
    "order_id",
    "payment_value"
]]

customers = customers[[
    "customer_id",
    "customer_city",
    "customer_state"
]]

products = products[[
    "product_id",
    "product_category_name"
]]
```

--------
Test joins
--------
--------

```python
df = orders.merge(order_items, on="order_id") \
           .merge(products, on="product_id") \
           .merge(customers, on="customer_id") \
           .merge(payments, on="order_id")

print(df.head())
print(df.shape)
```
